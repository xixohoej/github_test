```python
s=input()
count=1
result=0
for i in s:
    for j in range(2,len(s)+1):
        if s.find(i*j)!=-1:
            result=1
            if j>count:
                count=j
if result==1:
    print("包含")
else:
    print("不包含")
print("最长长度为："+str(count))
```

```python
import random
S = 100
N = 10000000
C = 0
for i in range(N):
    x = random.uniform(0.0,1.0)
    y = random.uniform(0,100)
    if y <= x*x*x+x*x:
        C += 1
I = C / N * S
print(I)
```

```python
def square_root_1():
    c=2
    i=0
    g=0
    for j in range(0,c+1):
        if(j*j>c and g==0):
            g=j-1
    while (abs(g*g-c)>0.0001):
        g+=0.00001
        i=i+1
        print("%d:g=%.5f"%(i,g))
square_root_1()
```

```python
def square_root_2():
    i=0
    c=2
    m_max=c
    m_min=0
    g=(m_min+m_max)/2
    while (abs(g*g-c)>0.00000000001):
        if (g*g<c):
            m_min=g
        else:
            m_max=g
        g=(m_min+m_max)/2
        i=i+1
        print("%d:g = %.13f" % (i,g))
        
square_root_2()
```

```python
def square_root_3():
    c=2
    g=c/2
    i=0
    while (abs(g*g-c)>0.00000000001):
        g=(g+c/g)/2
        i=i+1
        print("%d:%.13f"%(i,g))
        
square_root_3()
```
